### Clustering

```python
df_1 = df.copy()
df_1['Gender'] = df['Gender'].replace({'Male' : 0, 'Female' : 1})
df_1.drop('CustomerID', axis = 1, inplace = True) 
df_1.head()
```

```python
from sklearn.cluster import KMeans
inertia = []
for n in range(1 , 11):
    algorithm = (KMeans(n_clusters = n ,init='k-means++', n_init = 10 ,max_iter=300, 
                        tol=0.0001,  random_state= 111  , algorithm='elkan') )
    algorithm.fit(df_1)
    inertia.append(algorithm.inertia_)
```

```python
plt.figure(1 , figsize = (15 ,6))
plt.plot(np.arange(1 , 11) , inertia , 'o')
plt.plot(np.arange(1 , 11) , inertia , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('Inertia')
plt.show()
```

### No. of clusters = 6

```python
algorithm = (KMeans(n_clusters = 6 ,init='k-means++', n_init = 10 ,max_iter=300, tol=0.0001, random_state= 111, algorithm='elkan'))
algorithm.fit(df_1)
labels = algorithm.labels_
centroids = algorithm.cluster_centers_
```

```python
df_1['cluster'] = labels
df_1['cluster'] = df_1['cluster'].replace({0 : 'cluster_0', 1 : 'cluster_1', 2 : 'cluster_2', 3 : 'cluster_3', 4 : 'cluster_4', 5 : 'cluster_5'})
df_1.head()
```

```python
plt.figure(figsize = (16, 7))
ax = plt.axes() 

sns.scatterplot(
    x = df_1['Annual Income (k$)'],
    y = df_1['Spending Score (1-100)'],
    hue = df_1['cluster'],
    palette = ['#EE82EE', '#DC143C', '#3CB371', '#FF1493', '#4169E1', 'grey'],
    s = 150,
    alpha = 0.6,
    linewidth = 2
)

sns.despine()

plt.xlabel(
    xlabel = 'Annual Income (k$)',
    color = 'black',
    fontsize = 16,
    fontname = 'Arial'
)

plt.ylabel(
    ylabel = 'Spending Score (1-100)',
    color = 'black',
    fontsize = 16,
    fontname = 'Arial'
)

plt.xticks(
    range(0, 141, 10),
    fontsize = 14, 
    fontname = 'Arial', 
    color = 'black',
)

plt.yticks(
    range(0, 101, 10),
    fontsize = 14, 
    fontname = 'Arial', 
    color = 'black'
)

plt.legend(
    loc = 1,
    bbox_to_anchor=(0.72, 0.6, 0.5, 0.5),
    fontsize = 12,
    markerscale = 2,
    title = 'Segment',
    title_fontsize = 14)

plt.title(
    label = 'Clusters',
    loc = 'left',
    y = 1.1,
    x = - 0.09,
    color = 'black',
    fontname = 'Arial',
    fontsize = 24,
    fontweight = 'bold'
)

plt.show()
```
